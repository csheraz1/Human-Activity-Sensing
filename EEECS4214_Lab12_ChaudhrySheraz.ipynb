```python
import numpy as np
import h5py as h5
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
import scipy.io as io
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
x_test = np.load("/cs/home/csheraz1/CSI_Dataset/data/x_test.csv")
x_train = np.load("/cs/home/csheraz1/CSI_Dataset/data/x_train.csv")
y_test = np.load("/cs/home/csheraz1/CSI_Dataset/label/y_test.csv")
y_train = np.load("/cs/home/csheraz1/CSI_Dataset/label/y_train.csv")
## convert data into one dimension
x_test = x_test.reshape(x_test.shape[0], (x_test.shape[1]*x_test.shape[2]))
x_train = x_train.reshape(x_train.shape[0], (x_train.shape[1]*x_train.shape[2]))
#x_testSub = x_test[:,2,:]
#x_trainSub = x_train[:,2,:]
```

```python
#### K neighbors
# neighbors = 4 is highest accuracy for now when second dimeions is one. Maybe try for for multiple iterations and 
## store result
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
## Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("neighbors = 1 accruacy is ", accuracy)

knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

# Top performing case
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [1, 2, 3, 4, 5, 6, 7])
cm_display.plot()

## Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("neighbors = 2 accruacy is ",accuracy)

knn = KNeighborsClassifier(n_neighbors =5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
## Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("neighbors = 5 accruacy is ",accuracy)

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
## Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("neighbors = 10 accruacy is ",accuracy)

knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("neighbors = 10 accruacy is ",accuracy)
```

```python
#### SVM
clf = svm.SVC(kernel = 'linear')

clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [1, 2, 3, 4, 5, 6, 7])
cm_display.plot()
```

```python
#### Random forest
rf = RandomForestClassifier(n_estimators = 5)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for 5 estimators is ", accuracy)

rf = RandomForestClassifier(n_estimators = 10)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for 10 estimators is", accuracy)

rf = RandomForestClassifier(n_estimators = 15)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for 15 estimators is ", accuracy)

# Top performing case confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [1, 2, 3, 4, 5, 6, 7])
cm_display.plot()
```
